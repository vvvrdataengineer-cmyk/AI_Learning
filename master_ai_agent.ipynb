 Q. what is an AI agent?
 an AI agent is a system that can:
 1. receive instructions.
 2. think about the task.
 3. use tools if needed.
 4. perform an action.
 
 Input: user gives instruction.
  eg: find top competitors of tesla.
  
 Thinking: AI analyses the request.
 
 tools: AI may use tools like:
	internet search.
	database.
	email system.
	
Actions: AI perform required action and produces the result.

eg: report 
	email reply.
	summarized data.
	
So an AI agent is not just answering questions. it is performining tasks/actions.

flow:
receive user request.
understand the task.

execute the task.
return the result.

Interview Questions:
Q. what is Artificial intelligence?
- AI is the technology that enables machines to perform tasks that normally require human intelligence, such as recognizing patterns, understanding language, and making decisions.



Q. what is generative AI?
- is type of AI that can generate new text, images, code or music.

Q. what is the difference between AI and Generative AI?
- AI analyzes data and make decisions.

- Generative AI creates new content based on patterns learned from data.



What is agentic AI:
- AI systems that can take actions to complete tasks.
- They behave like digital agents/assistants.

An agents is something that can:
 observe a situation.
 think about the problem.
 take actions.
