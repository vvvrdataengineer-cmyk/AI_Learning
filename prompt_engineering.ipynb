**Prompt Formula:**

**RTC CO: Role + Task + Context + Constraints + Output**

Set Role.--> GIVES DIRECTION.

Set Task.--> GIVES WORK.

Set Context.--> GIVES CLARITY.

Set Constraints. --> GIVES CONTROL.

Set Output format.--> GIVE SHAPE.

**Role:**

 Who should AI act like?
 What role should AI play?
 
**Ex:**
**Act as a Teacher.**

**Act as a HR.**

**Act as a interviewer.**

**Act as a manager.**

**Act as a mentor.**
 
 Role gives direction to the AI.
 without role AI gives a general answer.
 
** Task: What should AI do?**
 what work you want to AI do.
 
 Ex:
 Explain python.
 Write email.
 Create resume.
 ask interview question.
 
 Task gives job to AI.
 
** 3. Context - For whom / background:**
 background information.
 
 Ex:
 For school students.
 For fresher.
 For working professional.
 For business owner.
 
 Context tells AI level and situation.

 **4. Constraints: Rules and Limits**
 Restrictions or conditions.
 
 Eg:
 Use simple english.
 do not use technical words.
 answer in 5 points.
 keep it short.
 
 Constraints control AI behaviour.
 
 **5. Output Format - How answer should look like:**
 How do you want the answer?
 
 Eg:
 Bullet points.
 Tables.
 steps.
 short paragraph.
 
 Format gives shape to the output.

**Prompt Engineering - 5 Key Points:**

1. Prompt engineering means giving clear instructions to AI to get correct and useful output.
2. Be specific and clear: a good prompt explains what to do, how to do, and expected result.
3. clear prompt reduce wrong answers and confusion from AI.
4. Changing words in a prompt can change the output completely.
5. Prompt engineering helps in writing content, code , emails, answers, and ideas faster.

**Tone Control:**

style of speaking.
mood of speaking.
how the answer should "feel"

**4 common tones:**

1. friendly: answer becomes soft and simple.
	soft, simple, caring, like a helpful friend.
	usage: begineers, school students, people who fear technical topics.
	
2. professional: answer becomes formal and office style.
	office style, formal, clean, polite, no jokes.
	usage: emails, reports, client messages, linkedin post.
	
3. funny: answer includes jokes and fun examples.
	jokes, light styple, fun examples, makes boring topic interesting.
	usage: reels, youtube shorts, marketing content.
	
4. strict: answer becomes direct and serious.
	direct, serious, no extra words, teaching or coaching.
    Usage: interview questions, quick revision, strong discipline style learning.
